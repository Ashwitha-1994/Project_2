```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)
# Select relevant columns for all 3 crops
crops = {
    "Rice": ["RICE AREA (1000 ha)", "RICE PRODUCTION (1000 tons)"],
    "Wheat": ["WHEAT AREA (1000 ha)", "WHEAT PRODUCTION (1000 tons)"],
    "Maize": ["MAIZE AREA (1000 ha)", "MAIZE PRODUCTION (1000 tons)"]
}
```

```python
plt.figure(figsize=(15, 5))

# Loop through each crop and make a scatter plot
for i, (crop, cols) in enumerate(crops.items(), 1):  #gives subplot index i starting at 1 (matplotlib expects subplot numbers starting at 1).
    area_col, prod_col = cols
    plt.subplot(1, 3, i) #creates the i-th subplot in a 1×3 grid (one row, three columns)
    sns.scatterplot(data=df1, x=area_col, y=prod_col, alpha=0.6, color="teal") #draws points: each point represents one row in the DataFrame (a district/region/year depending on your data).
    
    # Add regression line (trendline)
    sns.regplot(data=df1, x=area_col, y=prod_col, scatter=False, color="orange") #draws a linear regression line (least squares) on top of the scatter
    
    plt.title(f"{crop}: Area vs Production") #label the axes and plot.
    plt.xlabel("Area Cultivated (1000 ha)")
    plt.ylabel("Production (1000 tons)")
    plt.grid(alpha=0.3) #draws light gridlines for easier reading.

plt.suptitle("Impact of Area Cultivated on Production (Rice, Wheat, Maize)", fontsize=14,color='Red') #adds a big title for the whole figure.
plt.tight_layout(rect=[0, 0, 1, 0.96]) #prevents labels/titles from overlapping
plt.show() 
```

```python
#correlation
for crop, cols in crops.items():
    corr = df1[cols[0]].corr(df1[cols[1]])
    print(f"{crop} correlation between area and production: {corr:.2f}")
```
